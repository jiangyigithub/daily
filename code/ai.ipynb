```python
import matplotlib.pyplot as pyplot

X=[.01*i for i in range(100)]
Y= [2*x**2+3*x+4 for x in X] ## Groud truth
print(X)
print(Y)
pyplot.scatter(X,Y)
pyplot.show()
```

```python
# model
def fun1(x):
    return w1*x**2+w2*x+w3

# 损失函数
def loss(y_p,y_t):
    return (y_p-y_t)**2

# 权重（weight）
w1,w2,w3 =-1,1,0

# 学习率
lr=0.01 

for epoch in range(10000):
    epoch_loss =0
    for x,y_true in zip(X,Y):
        y_pred = fun1(x)
        epoch_loss+=loss(y_pred,y_true)
        # 梯度下降
        grad_w1 =2*(y_pred-y_true)*x**2
        grad_w2=2*(y_pred-y_true)*x
        grad_w3=2*(y_pred-y_true)
        
        # 更新权重（weight）--> learning
        w1=w1-lr*grad_w1
        w2=w2-lr*grad_w2
        w3=w3-lr*grad_w3
    epoch_loss/=len(X)
    print(f'at {epoch} round,loss:{epoch_loss}')
    if epoch_loss <0.01:
        break
print(f'after training,w1:{w1};w2:{w2}:w3:{w3}')
Y1=[fun1(i) for i in X]
```

```python
import matplotlib.pyplot as pyplot
pyplot.scatter(X,Y,color="red")
pyplot.scatter(X,Y1)
pyplot.show()
```
